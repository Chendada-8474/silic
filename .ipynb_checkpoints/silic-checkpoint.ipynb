```python
%cd 'silic'
!git pull
import pandas as pd
import silic
```

```python
model = silic.Silic()
model.audio('sample/49593.mp3')
```

```python
%matplotlib inline
model.tfr()
```

```python
%matplotlib inline
model.tfr(spect_type='mel')
```

```python
%matplotlib inline
model.tfr(spect_type='rainbow')
```

```python
labels = model.detect(weights='model/exp12/best.pt', step=1000, soundclasses=[66,101], conf_thres=0.1)
```

```python
df = pd.DataFrame(labels[1:],columns=labels[0])
df = df.sort_values('time_begin')
print(df)
```

```python
newlabels = silic.clean_multi_boxes(labels)
```

```python
print(newlabels)
```

```python
silic.draw_labels(model, newlabels, 'sample')
```
